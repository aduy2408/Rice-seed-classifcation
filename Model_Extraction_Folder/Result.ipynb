```python
df = df.rename(columns={'Feature Combination':'Feature','Dataset':'Rice Seed','F1 Score':'F1-score'})
```

```python
feature = df["Feature"].unique()
feature
```

```python
group_feature = df.groupby("Feature")
```

```python
group_feature.get_group("Basic+Enhanced Color+Zernike moments+Edge")
```

```python
def create_result(group, feature_type):
    fea = group.get_group(feature_type).pivot_table(index=["Rice Seed"], columns=["Model"], values="Accuracy")
    tmp_df = pd.DataFrame(fea)
    mean_vals = fea.loc[:, :].mean()
    mean_mdl = pd.DataFrame([mean_vals.tolist()], columns=fea.columns, index=["AVERAGE"])
    return pd.concat((tmp_df, mean_mdl), axis=0)
```

```python
create_result(group_feature, "Basic+Enhanced Color+Zernike moments+LBP+GLCM+Edge")
```
